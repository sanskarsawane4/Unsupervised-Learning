```python
import pandas as pd
```

```python
data=pd.read_csv("Mall_Customers.csv")
```

```python
data
```

```python
data.columns
```

```python
x=data.iloc[:,[3,4]].values
```

```python
x
```

```python
import scipy.cluster.hierarchy as sc
dendrogram=sc.dendrogram(sc.linkage(x,method="ward"))
```


we will find the maximum vertical distance that does not cut any horizontal bar. as this graph shows that best suitable number of clusters is 5

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler.fit_transform(x)
```

```python
from sklearn.cluster import AgglomerativeClustering
clusterer=AgglomerativeClustering(n_clusters=5,affinity="euclidean",linkage="ward")
y_pred=clusterer.fit_predict(x)
```

```python
y_pred
```

```python
!pip install matplotlib
```

```python
import matplotlib.pyplot as mtp
mtp.scatter(x[y_pred == 0, 0], x[y_pred == 0, 1], s = 100, c = 'blue', label = 'Cluster 1')
mtp.scatter(x[y_pred == 1, 0], x[y_pred == 1, 1], s = 100, c = 'green', label = 'Cluster 2')
mtp.scatter(x[y_pred== 2, 0], x[y_pred == 2, 1], s = 100, c = 'red', label = 'Cluster 3')
mtp.scatter(x[y_pred == 3, 0], x[y_pred == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
mtp.scatter(x[y_pred == 4, 0], x[y_pred == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
mtp.title('Clusters of customers')
mtp.xlabel('Age')
mtp.ylabel('Spending Score (1-100)')
mtp.legend()
mtp.show()
```

We have successfully divided the customers based on their Age and Annual Income
